```python
#Importing Pandas
import pandas as pd
#Importing Numpy
import numpy as np
#Importing Matplotlib
import matplotlib.pyplot as plt

```

```python
#Loading CSV File
data = pd.read_csv('/content/drive/MyDrive/householdtask3.csv')
#Displaying the first 10 rows of the CSV File
data.head(10)
```

```python
#Scatter Plot with Year against income
plt.scatter(data['year'],data['income'])

#Adding Title to the plot
plt.title('Scatter Plot')

#Inserting X and Y Labels
plt.xlabel('year')
plt.ylabel('income')

#To display the plot
plt.show()
```

```python
#Line Chart with Year against income
plt.plot(data['year'])
plt.plot(data['income'])

#Adding Title to the Chart
plt.title('Line Chart')

#Inserting X and Y Labels
plt.xlabel('year')
plt.ylabel('income')

#To display the Chart
plt.show()
```

```python
#Bar Chart with Year against income
plt.bar(data['year'],data['income'])

#Adding Title to the plot
plt.title('Bar Plot')

#Inserting X and Y Labels
plt.xlabel('year')
plt.ylabel('income')

#To display the plot
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

#Adding Title
plt.title('Histogram')

#Displaying
plt.show()
```
